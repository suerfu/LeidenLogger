```python
# This python script is used to plot the contour map of chi2 obtained for the fit
# of zone refining data

# input is supposed to be a three-column, comma-separated raw data
# first column is the segregation coefficient k
# second column is the estimated initial concentration C
# third column is the chi square of the fit using the corresponding two parameters.

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib.ticker import MultipleLocator
import pandas as pd

plt.rcParams["font.family"] = "Times New Roman"
```

```python
data=pd.read_csv( 'run15_demo_20210526_liqlev.txt', header=None, comment='#')
data=data.values
plt.figure(figsize=(20,16))
ax = plt.axes()
plt.plot( data[:,0]/3600, data[:,1], label='Liquid helium level', color='r', linewidth=3)

plt.xlabel('Hours since 5/26/2021, 8:38:44 PM', fontsize=32);
plt.ylabel('Liquid helium level', fontsize=32);

ax.minorticks_on()
ax.tick_params('both', length=20, width=1, which='major')
ax.tick_params('both', length=10, width=1, which='minor')

plt.xticks(fontsize=32)
plt.yticks(fontsize=32)

#plt.xlim([0,1])

plt.grid(True, which='both')

plt.legend(fontsize=34)
```

```python
data=pd.read_csv( 'run15_demo_20210526_pres.txt', header=None, comment='#')
data=data.values
plt.figure(figsize=(20,16))
ax = plt.axes()
plt.plot( data[:,0]/3600, data[:,1], label='Condenser', linewidth=3)
plt.plot( data[:,0]/3600, data[:,2], label='Still', linewidth=3)
plt.plot( data[:,0]/3600, data[:,3], label='Dump', linewidth=3)
plt.plot( data[:,0]/3600, data[:,4], label='Pot', linewidth=3)
plt.plot( data[:,0]/3600, data[:,5], label='IVC', linewidth=3)

plt.yscale('log')

plt.xlabel('Hours since 5/26/2021, 8:38:44 PM', fontsize=32);
plt.ylabel('Pressure (mbar)', fontsize=32);

ax.minorticks_on()
ax.tick_params('both', length=20, width=1, which='major')
ax.tick_params('both', length=10, width=1, which='minor')

plt.xticks(fontsize=32)
plt.yticks(fontsize=32)

#plt.xlim([0,1])

plt.grid(True, which='both')

plt.legend(fontsize=34)
```
